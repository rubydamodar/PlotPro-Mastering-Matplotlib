## Basic Line Plot 

- 1. Basic Line Plot 
    
    here i am creating a basic line plot using a list of value x and y 

```python
import matplotlib.pyplot as plt
```

imagine you've drawing a simple graph on pape you have few number for X-axis and Y-axis like youre class marks for different subjects lets plot them 

```python
x=[1,2,3,4]
y=[50,80,70,90]
plt.plot(x,y)
plt.show()
```

here we're just drawing a simple line connecting marks 

## Adding Colors 

we always like things a bit colors lets add some color to our line !

```python
plt.plot(x,y,color='green')
plt.show()
```

## Multiple lines ( comparing different subjects)

now lets say you want to compare youre marks from this year with last year you'll plot two lines - one for this year and one for last year

```python
y_last_year = [40,60,75,85] # Marks From Last Year
plt.plot(x,y,color='red',label='This Year')
plt.legend() 
plt.show()
```

> ### `legend()`= to show which line is which 

## Changing Line Style ( Adding Variety)

sometimes often we want to make one line dashed or dotted to differentiate them batter , line in cricket we have different types of bowlers ( fast , spin , etc...)

```python
plt.plot(x,y,color='green' , linestyle = '--', label='This Year')   
plt.plot(x,y_last_year , color='red',linestyle='-.',label='Last Year') 
plt.legend()
plt.show()
```

## Adding Markers (Highlight Points)

if you want to highlight each subjects mark (just like putting a dot on a paper),you can use markers think of it as adding points where the line passess through 

```python
plt.plot(x,y,marker='o', color='blue',label='This Year')
plt.plot(x,y_last_year,marker='x',color='red',label='Last Year')
plt.legend()
plt.show()
```

Customizing the graph ( making it look professionals)

## Adding Labels ( Naam To Zaroori hai mittr 😂)

just like we give names to our cricket team player or movie stars , we should name our graph's X-ais and Y-axuis.

```python
plt.plot(x,y,color='green',label='This Year')
plt.xlabel('Subjects') # Name for The X-axis
plt.ylabel('Marks') # Name for the y-axis
plt.title('Marks Comparision') # Graph Title 
plt.show()
```

## Grid (setting the stage)

gridlies make the graph look clearer , just like a cricket field is divided into different areas

```python
plt.plot(x,y,color='blue')
plt.grid(True)
plt.show()
```

## Multiple (Lines Compare and contracts)

when we compare marks , we can have many years on the some graphs lets add one more year !

```python
y_two_years_ago=[35,75,65,80]
plt.plot(x,y,color='blue',label='This Year')
plt.plot(x,y_last_year,color='red',label='Last Year')
plt.plot(x,y_two_years_ago,color='green',label='Two Years Ago')
plt.legend()
plt.show()
```

# Lets Change Line width 

like how bold and light fonts can make a difference in reading you can change the thickness of lines to highlight important ones...

```python
plt.plot(x,y,color='red',linewidth=4,label='This Year') # Thicckness of line 
plt.plot(x,y_last_year,color='red',linewidth=1,label='Lat Year') # Thinner Line 
plt.legend()
plt.show()
```

## Subplots (Arranging Graphs Like A Gallery)

let's say you want to show multiple graph in the same window just like when we hang different photos in our house

```python
fig,(ax1,ax2) = plt.subplots(2,1) # here two subplots in a columns
ax1.plot(x,y,color='blue',label='This Year')
ax2.plot(x,y_last_year,color='red',label='Last Year')
plt.show()
```

# Fill Area Under The Curve 

we can fill the area under the curve , like coloring the space below a line 

```python
plt.plot(x,y,color='blue')
plt.fill_between(x, y , color='lightblue') # filling the area
plt.show()
```

## Logarithm Scale its tricky but its important

sometimes we use log scales when data grows exponentially ( like our population ).

```python
plt.plot(x,y)
plt.yscale('log')
plt.show()
```

## Dynamic Range ( Control The view)

we can control what part of the graph to zoom into just like focusing a camera lens 

```python
plt.plot(x,y,color='blue')
plt.xlim(2,4) # x-axis 2 to 4
plt.ylim(60,100) # y-axis from 60 to 100
plt.show()
```

## Line Styles In Detaiuls (like a flag )

there are many ways to change the style of line we already know about dashed and dotted lines

```python
plt.plot(x,y,color='green',linestyle='-',label='Solid Line') # Default Solid Line
plt.plot(x,y_last_year,color='red',linestyle='--',label='Dashed Line')
plt.plot(x,y_two_years_ago,color='blue',linestyle=':')
plt.legend()
plt.show()
```
